```python
import random
from datetime import datetime, timedelta
import pandas as pd


def generate_exact_survey_data_with_timestamp(num_rows=1800):
    # Precise MCQ options extracted from the provided form content
    employment_options = ["Employed", "Self-employed", "Unemployed", "Student"]
    education_options = ["Undergraduate", "Postgraduate", "Vocational", "Other"]
    age_options = ["18-25", "26-35", "36-45", "46-55", "56 - above"]
    income_options = [
        "Below 50,000",
        "50,000-100,000",
        "100,001-200,000",
        "200,001-500,000",
        "Above 500,000",
    ]
    bulk_food_options = ["Always", "Sometimes", "Rarely", "Never"]
    food_pct_options = ["Less than 20%", "20-40%", "41-60%", "more than 60%"]
    food_reduction_options = [
        "Yes, both quantity and quality",
        "Yes, only quantity",
        "Yes, only quality",
        "No",
    ]
    food_bill_increase_options = [
        "Less than 5,000",
        "5,000-15,000",
        "15001-30,000",
        "More than 30,000",
    ]
    transport_spend_options = [
        "Below 10,000",
        "10,000-20,000",
        "20,001-30,000",
        "Above 30,000",
    ]
    transport_mode_options = [
        "Private car/bike/bicycle",
        "Bike",
        "Public bus",
        "Walking",
    ]
    fare_increase_options = [
        "Less than 10%",
        "10-25%",
        "26%-50%",
        "More than 50%",
    ]
    move_house_options = ["Never", "Once", "Twice", "More than twice"]
    power_cut_options = [
        "Yes, frequently",
        "Yes, occasionally",
        "Rarely",
        "Never",
    ]
    rent_pct_options = ["Less than 20%", "20-35%", "36-50%", "More than 50%"]
    rent_increased_options = ["Yes", "No", "I own my home"]
    skipped_medical_options = ["Yes, more than once", "Yes, once", "No"]
    insurance_options = ["Yes", "No"]
    burden_options = [
        "Food",
        "Rent",
        "Transportation",
        "Healthcare",
        "Education",
        "Utilities",
    ]
    future_increase_options = [
        "Definitely yes",
        "Probably yes",
        "Uncertain",
        "Probably no",
        "Definitely no",
        "No",
    ]
    coping_pool = [
        "Reduced spending",
        "Taking a second job",
        "Borrowed money",
        "Relocated to a cheaper area",
        "Started a small business",
    ]
    factor_options = [
        "Inflation (rising prices)",
        "Fuel prices",
        "Currency depreciation",
        "Government policies",
        "Other",
    ]
    year_options = [
        "Before 2018",
        "2018",
        "2019",
        "2020",
        "2021",
        "2022",
        "2023",
        "2024",
        "2025",
        "2026",
    ]

    cameroon_cities = [
        "Douala",
        "Yaounde",
        "Bamenda",
        "Bafoussam",
        "Garoua",
        "Maroua",
        "Buea",
        "Kribi",
    ]
    food_items = [
        "Rice",
        "Cooking oil",
        "Onions & Tomatoes",
        "Plantains / Cassava",
        "Fish / Meat",
        "Bread",
    ]

    # Baseline start time for the survey submissions (simulate responses over a 5-day window)
    start_date = datetime(2026, 5, 26, 8, 0, 0)

    data = []

    for i in range(1, num_rows + 1):
        income = random.choice(income_options)

        # Generate a realistic rolling timestamp
        # Increments each submission by a random spread of minutes and seconds
        start_date += timedelta(
            minutes=random.randint(1, 15), seconds=random.randint(0, 59)
        )
        timestamp_str = start_date.strftime("%Y/%m/%d %I:%M:%S %p")

        if "Below 50,000" in income:
            base_inc = 40000
        elif "50,000-100,000" in income:
            base_inc = 75000
        elif "100,001-200,000" in income:
            base_inc = 150000
        elif "200,001-500,000" in income:
            base_inc = 350000
        else:
            base_inc = 650000

        food_spend_xaf = int(base_inc * random.uniform(0.3, 0.55))
        rent_spend_xaf = (
            0
            if random.random() < 0.15
            else int(base_inc * random.uniform(0.2, 0.4))
        )
        utilities_xaf = random.randint(3000, 25000)
        healthcare_xaf = random.randint(2000, 35000)
        school_fees_xaf = random.choice([0, 35000, 75000, 150000, 300000])

        num_measures = random.randint(1, 3)
        chosen_measures = random.sample(coping_pool, k=num_measures)
        coping_measures_string = "; ".join(chosen_measures)

        row = {
            "Timestamp": timestamp_str,
            "Q1_Employment_Status": random.choice(employment_options),
            "Q2_Highest_Education": random.choice(education_options),
            "Q3_Age_Group": random.choice(age_options),
            "Q4_City_Residence": random.choice(cameroon_cities),
            "Q5_Monthly_Household_Income_XAF": income,
            "Q6_Household_Size_Supported": random.randint(1, 8),
            "Q7_Bulk_Food_Frequency": random.choice(bulk_food_options),
            "Q8_Percentage_Income_To_Food": random.choice(food_pct_options),
            "Q9_Reduced_Food_Quality_Quantity": random.choice(
                food_reduction_options
            ),
            "Q10_Monthly_Food_Spend_XAF_Numeric": food_spend_xaf,
            "Q11_Estimated_5Yr_Food_Bill_Increase": random.choice(
                food_bill_increase_options
            ),
            "Q12_Most_Dramatic_Increased_Food_Item": random.choice(food_items),
            "Q13_Monthly_Transport_Spend": random.choice(
                transport_spend_options
            ),
            "Q14_Fuel_Prices_Impact_Scale_1_5": random.randint(3, 5),
            "Q15_Primary_Transportation_Mode": random.choice(
                transport_mode_options
            ),
            "Q16_Paid_Transport_Trips_Per_Week": random.randint(0, 14),
            "Q17_Estimated_5Yr_Transport_Fare_Increase_Percent": random.choice(
                fare_increase_options
            ),
            "Q18_Times_Moved_Due_To_Rent": random.choice(move_house_options),
            "Q19_Power_Cuts_Extra_Spend": random.choice(power_cut_options),
            "Q20_Percentage_Income_To_Rent": random.choice(rent_pct_options),
            "Q21_Monthly_Rent_XAF_Numeric": rent_spend_xaf,
            "Q22_Monthly_Utilities_Combined_XAF_Numeric": utilities_xaf,
            "Q23_Rent_Increased_Last_2_Years": random.choice(
                rent_increased_options
            ),
            "Q24_Skipped_Medical_Treatment_Cost": random.choice(
                skipped_medical_options
            ),
            "Q25_Schooling_Cost_Increased_Scale_1_5": random.randint(3, 5),
            "Q26_Has_Health_Insurance": random.choice(insurance_options),
            "Q27_Average_Monthly_Healthcare_Spend_XAF_Numeric": healthcare_xaf,
            "Q28_School_Fees_Per_Child_XAF_Numeric": school_fees_xaf,
            "Q29_Highest_Burden_Expenditure_Category": random.choice(
                burden_options
            ),
            "Q30_Expect_Cost_Of_Living_Increase_5Yrs": random.choice(
                future_increase_options
            ),
            "Q31_Taken_Measures_To_Cope": coping_measures_string,
            "Q32_Primary_Contributor_To_Cost_Of_Living": random.choice(
                factor_options
            ),
            "Q33_Year_Noticeable_Impact": random.choice(year_options),
        }
        data.append(row)

    df = pd.DataFrame(data)

    # Saved with a brand new name ('_v3.csv') to prevent any Excel lock errors
    output_filename = "cameroon_cost_of_living_survey_v3.csv"
    df.to_csv(output_filename, index=False, encoding="utf-8")
    print(
        f"Success! Generated {df.shape[0]} rows and saved to '{output_filename}'."
    )
    return df


# Run script
df_preview = generate_exact_survey_data_with_timestamp(1800)
df_preview.head()
```
